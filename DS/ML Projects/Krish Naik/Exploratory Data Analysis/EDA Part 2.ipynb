```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
pd.pandas.set_option("display.max_columns",None)
```

```python
df=pd.read_csv("HousePrice.csv")
df.head()
```

```python
numericals=[feat for feat in df.columns if df[feat].dtypes!='O'] 
Dat_lst=[feat for feat in df.columns if 'Yr' in feat or 'Year' in feat]
disc_feat=[feat for feat in numericals if len(df[feat].unique())<25 and feat not in Dat_lst+['Id']]
```

```python
cont_feats=[feat for feat in numericals if feat not in disc_feat + ['Id']]
cont_feats
```

```python
for feat in cont_feats:
    df[feat].hist(bins=25)
    plt.title(feat)
    plt.show()
```

### transforming above skewd feats to normal distrib

```python
cat_feats=[feat for feat in df.columns if df[feat].dtypes=='O']
cat_feats
```

```python
for feat in cat_feats:
    temp=df.groupby(feat)['SalePrice'].count()/len(df)
#     print(temp)
    print(feat,(temp[temp>0.01]))
```

```python
for feat in cat_feats:
    temp=df.groupby(feat)['SalePrice'].count()
    print(df[feat].isin(temp[temp>0.01].index))
```

```python
for feat in cat_feats:
    temp=df.groupby(feat)['SalePrice'].count()
    
    print(feat,(temp/len(df)))
```
