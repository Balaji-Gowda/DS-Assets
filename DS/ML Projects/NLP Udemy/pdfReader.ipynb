# Reading

```python
import PyPDF2
```

```python
fil=open('Excel_answers.pdf',mode='rb')
```

```python
pdf_reader=PyPDF2.PdfFileReader(fil)
```

```python
pdf_reader.numPages
```

```python
pg1=pdf_reader.getPage(0)
```

```python
mytext=pg1.extractText()
```

```python
print(mytext)
```

# writing 

```python
f=open('Excel_answers.pdf','rb')
pdf_re=PyPDF2.PdfFileReader(f)
fstpage=pdf_re.getPage(0)
pdf_writer=PyPDF2.PdfFileWriter()
pdf_writer.addPage(fstpage)
```

```python
pdf_op=open('NewPdfWrite.pdf','wb')
```

```python
pdf_writer.write(pdf_op)
pdf_op.close()
f.close()
```

```python
p1=open('NewPdfWrite.pdf','rb')
p2=PyPDF2.PdfFileReader(p1)
p3=p2.getPage(0)
p4=p3.extractText()
print(p4)
```
