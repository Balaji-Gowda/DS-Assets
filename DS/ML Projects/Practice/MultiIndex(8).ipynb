# The swapLevel Method

# The pivot_table() method

```python
foods=pd.read_csv("foods.csv")
foods.head()
```

```python
foods.info()
```

```python
foods.pivot_table(values="Spend",index="Gender",aggfunc="mean")
```

```python
foods.pivot_table(values="Spend",index="Item",aggfunc="sum")
```

```python
foods.pivot_table(values="Spend",index="Gender",aggfunc="sum")
```

```python
(foods.pivot_table(values="Spend",index=["Item","Gender","City"],aggfunc="sum"))
```

```python
foods.pivot_table(values="Spend",index=["Gender","Item","City"],aggfunc="sum")
```

```python
foods.pivot_table(values="Spend",index=["Gender","Item"],columns="City",aggfunc="sum").head()
```

```python
foods.pivot_table(values="Spend",index=["Gender","Item"],columns=["Frequency","City"],aggfunc="sum")
```

```python
foods.pivot_table(values="Spend",index=["Gender","Item"],columns="City",aggfunc="max")
foods.pivot_table(values="Spend",index=["Gender","Item"],columns="City",aggfunc="max")
```

```python
#pivot table directly on pandas without double quotes for df name
pd.pivot_table(data=foods,values="Spend",index=["Gender","Item"],columns="City",aggfunc="max")
```
