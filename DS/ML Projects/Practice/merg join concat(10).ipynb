# outer join part-1

```python
wk1=pd.read_csv("Restaurant - Week 1 Sales.csv")

wk2=pd.read_csv("Restaurant - Week 2 Sales.csv")

cust=pd.read_csv("Restaurant - Customers.csv")

food=pd.read_csv("Restaurant - Foods.csv")
```

```python
wk1.merge(wk2,how="outer",on="Customer ID")
```

```python
wk1.merge(wk2,how="outer",on="Customer ID",indicator=True)#indicator
```

```python
mergeOut=wk1.merge(wk2,how="outer",on="Customer ID",indicator=True)#indicator
```

```python
mergeOut[mergeOut["_merge"]=="both"]#1          ###inner join only rows in both from outer join
```

```python
wk1.merge(wk2,how="inner",on="Customer ID")     ###same as above
```

```python
mergeOut[mergeOut["_merge"]!="both"].head()   ### only in left and only in right
```

```python
mergeOut[mergeOut["_merge"]!="both"].tail()   ### only in left and only in right
```
