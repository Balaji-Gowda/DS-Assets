```python
#one hot encoding for created months
df_created_months=pd.get_dummies(df['Created At'].dt.month_name(),drop_first=True)
df_created_months
```

```python
#one hot encoding for created days
df_created_days=pd.get_dummies(df['Created At'].dt.day,drop_first=True)
df_created_days
```

```python
#one hot encoding for Video created months
df_vid_create_months=pd.get_dummies(df['Video creation date'].dt.month_name(),drop_first=True)
df_vid_create_months
```

```python
#one hot encoding for Video created days
df_vid_create_days=pd.get_dummies(df['Video creation date'].dt.day,drop_first=True)
df_vid_create_days
```

```python
#one hot encoding for Video created days
df_vid_create_hours=pd.get_dummies(df['Video creation date'].dt.hour,drop_first=True)
df_vid_create_hours
```

```python
#one hot encoding for Video created days
df_vid_create_mins=pd.get_dummies(df['Video creation date'].dt.minute,drop_first=True)
df_vid_create_mins
```

```python
OH_encoded_df=pd.concat([df['ID'],df_created_months,df_created_days,df_vid_create_months,df_vid_create_days,df_vid_create_hours,df_vid_create_mins],axis=1)
OH_encoded_df.head()
```

```python
OH_encoded_df.shape
```

```python
#columns after one hot encoding
list(OH_encoded_df.columns)
```

```python
#reducing dimensions using PCA
from sklearn.decomposition import PCA
```

```python
OH_encoded_df.iloc[:,1:].describe()
```

```python
pca=PCA(n_components=15)
```

```python
df_dim_reduced=pca.fit_transform(OH_encoded_df.iloc[:,1:])
```

```python
# shape has been reduced to 15 from 150 columns
df_dim_reduced.shape
```

```python
df_dim_reduced=pd.DataFrame(df_dim_reduced)
```

```python
from sklearn.cluster import KMeans
```

```python
#finding no of clusters
wcss=[]
wcss2=[]
for i in range(1,15):
    k1=KMeans(n_clusters=i)
    m1=k1.fit(df_dim_reduced)
    wcss.append(k1.inertia_)
    wcss2.append(m1.inertia_)
```

```python
plt.plot(range(1,15),wcss,'p-')
```

```python
# from above fig it is clear that at 4, we have sharp break point which tells us 4 no of clusters to be chosen
```

```python
#using KMeans to cluster the indep variables into four clusters
k4=KMeans(n_clusters=4)
```

```python
m4=k4.fit(df_dim_reduced)
```

```python
clusters=m4.predict(df_dim_reduced)
```

```python
clusters
```

```python
df
```

```python
clustered_df=pd.concat([df,pd.DataFrame(clusters,columns=["predicted_cluster"])],axis=1)
clustered_df
```

```python
clustered_df.drop("index",axis=1,inplace=True)
```
