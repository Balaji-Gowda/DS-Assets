# 3) 1 - many relationship between planes - flights table

```python
manf=dict(list(zip(planes.tailnum,planes.manufacturer)))
```

```python
flights['manf']=flights.tailnum.map(manf)
```

```python
flights.head()
```

```python
flights['tot_delay']=flights['dep_delay']+flights['arr_delay']
flights.head()
```

```python
flights.head(30)
```

```python
f1=flights.copy()
f1.head()
```

```python
flights.head()
```

```python
f2=f1
```

```python
f2=f2.set_index('manf')
```

```python
f3=f2[['dep_delay','arr_delay']]
```

```python
f3.head(20)
```

```python
f3.shape
```

```python
list(((pd.DataFrame((f3.where(f3>0,0)['dep_delay']+f3.where(f3>0,0)['arr_delay']),columns=['total_delay'])).reset_index()).groupby('manf').total_delay.agg(['sum']).sort_values(by='sum',ascending=False).reset_index().loc[0])
```

```python
flights.groupby('manf').tot_delay.agg(['sum']).sort_values(by='sum',ascending=False).iloc[0,:]
```
