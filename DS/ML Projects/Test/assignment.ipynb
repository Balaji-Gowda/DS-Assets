```python
import pandas as pd
```

```python
df=pd.read_excel("Hair_Care_dataset.xlsx")
```

```python
df.head()
```

```python
df.shape
```

```python
df["Title"]
```

```python
import re
```

```python
def func(val):
    res=re.findall("\d{2,4}\s*Ml",val)
    if len(res)==0:
        return str(0)
    else:
        return res[0]
    
```

```python
vol=df['Title'].apply(func)
```

```python
vol
```

```python
df.insert(df.columns.get_loc("Title")+1,"volume in ML",vol)
```

```python
df.iloc[:,0:12]
```

```python
def repl(val):
    res=re.findall("\d{2,4}",str(val))
    if len(res)==0:
        return 0
    else:
        return res[0]
    
```

```python
df['volume in ML']=df['volume in ML'].apply(repl)
```

```python
df['volume in ML']
```

```python
df_inp=pd.DataFrame(df['volume in ML'])
```

```python
df_inp
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5)
```

```python
mod=km.fit(df_inp)
```

```python
pred=mod.predict(df_inp)
```

```python
output_df=pd.concat([df,pd.DataFrame(pred,columns=["Predicted cluster"])],axis=1)
```

```python
import seaborn as sns
```

```python
sns.pairplot(output_df,hue="Predicted cluster")
```

```python
output_df.to_excel("clustered output.xlsx",sheet_name="Clusters")
```
