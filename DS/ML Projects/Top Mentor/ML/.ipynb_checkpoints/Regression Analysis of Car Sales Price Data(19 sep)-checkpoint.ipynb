```python
import pandas as pd
df = pd.read_csv('Car_sales.csv')
df.head()
```

```python
df.shape
```

```python
#### Prediction of Car Sales based on Differnt Parameters - 
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df1 = df.dropna(axis = 0)
```

```python
df1.shape
```

```python
### All the preprocessing and Exploration has to be done first and then with the problem
#### Model has to be build
```

```python
y = df1['Sales_in_thousands']
X = df1[['Horsepower','Power_perf_factor','Fuel_capacity','Curb_weight']]
```

### Feature Selection

```python
#### Process of Feature Selection comes into picture
#### 1. Correlations
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot = True)
```

```python
### 2. Variance Selection Method - It helps to select the variables whose variance are high
df.var()
```

```python
X.shape,y.shape
```

### Sampling of the Datasets

```python
#### Sampling of the data // Dividing the dataset into Training and Testing
#### Scikit Learn - Module
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2)
```

```python
X_train.shape,y_train.shape
```

```python
X_train.head()  #### Randomly Divided to reduce the bias in the model
```

### Model Training

```python
#### Linear Regression - 1D and 1 ID or Multiple Regression - 1D, 2 ID
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()  #### y = m1x1 + m2x2 + c
```

```python
model_train = model.fit(X_train,y_train)  #### Model Training with the Training Dataset
print("The model is trained and Created")
```

```python
### Check the score of the model
model_train.intercept_   #### Training Data Intercept (a)
```

```python
model_train.coef_
```

```python
X1 = 45
X2 = 400
```

```python
### Rules
y = (2.66576999*X1 - -6.34056712*X2 +  49.05957162452773)
print("The Predicted Sales ",y)
```

```python
####Validate your Model
pred = model_train.predict(X_test)  #### 20 % of the data
pred
```

```python
compare = pd.DataFrame({'Actual_sales':y_test, 'Predicted_sales': pred,'Errors': (y_test - pred)})
compare
```

```python
m = model_train.score(X_train,y_train)
print(f"The accuracy of the model is {round(m*100,2)} %")
```
