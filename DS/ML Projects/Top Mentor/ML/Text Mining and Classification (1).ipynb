```python
### Stemming and Lemmitatization
#It is process of normalising the words into its base form or root word
### waited, waits, waiting - wait
There are two types o stemming - Porter Stemmer and Lancaster Stemmer(hasrh)
```

```python
from nltk.stem import PorterStemmer
pot = PorterStemmer()
pot.stem('considered')
```

```python
words = ['giving','given','gave','given']
for i in words:
    print(pot.stem(i))
```

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
```

```python
words = ['giving','given','gave','given']
for i in words:
    print(lst.stem(i))
```

```python
### Lemmitization - It is the process o converting a word to its base form
Lemmitization will correctly identiy the base form o caring to care, whereas 
stemming would cut off ing part and convert to car.
```

```python
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()
```

```python
words = ['step','steps','steeps','stepping']
for i in words:
    print(lm.lemmatize(i))
#print(lm.lemmatize('steps'))
```
