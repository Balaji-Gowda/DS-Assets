```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import datetime as dt
```

```python
df=pd.read_excel("Sensor Data Analysis.xlsx")
df["Time"][0]
```

```python
df["Time_Stamp"][0]
```

```python
#timestamp
type(df["Time_Stamp"][0])
```

```python
#timestamp to str
df['Time_Stamp']=df.Time_Stamp.dt.strftime('%Y-%m-%d')
df["Time_Stamp"][0]
print(df['Time_Stamp'][0])
type(df["Time_Stamp"][0])
```

```python
#time to string
print(type(df["Time"][0]))
df['Time']=df['Time'].astype(str)
print(type(df["Time"][0]))
```

```python
df.insert(0,"DateTime",df["Time_Stamp"]+" "+df["Time"])
df.head()
```

```python
type(df["DateTime"][0])
```

```python
df["DateTime"]=pd.to_datetime(df["DateTime"])
```

```python
type(df["DateTime"][0])
```

```python
df_index=df.set_index("DateTime")
```

```python
df_index.asfreq("5min").dropna(how="all")
```

```python
df_index.asfreq("30min").dropna(how="all")["F1_rHydPump_400Bar"].plot()
```
