```python
top_10_x1=[x for x in df['X1'].value_counts().head(10).index]
top_10_x3=[x for x in df['X3'].value_counts().head(10).index]
top_10_x4=[x for x in df['X4'].value_counts().head(10).index]
top_10_x5=[x for x in df['X5'].value_counts().head(10).index]
top_10_x6=[x for x in df['X6'].value_counts().head(10).index]
```

```python
#replacing top10 rows with 1s and remaining with 0s

def one_hot_encoding(data,col,top10):
    for val in top10:
        df[col+'_'+val]=np.where(data[col]==val,1,0)
```

```python
df.head()
```

```python
one_hot_encoding(df,'X2',top_10_x2)
```

```python
one_hot_encoding(df,'X1',top_10_x2)
one_hot_encoding(df,'X3',top_10_x2)
one_hot_encoding(df,'X4',top_10_x2)
one_hot_encoding(df,'X5',top_10_x2)
one_hot_encoding(df,'X6',top_10_x2)
```
