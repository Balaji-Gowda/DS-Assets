# The .between() mtd(Inclusive of both bounds)

# The .duplicated() mtd on Series

```python
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time",])
df["Gender"]=df["Gender"].astype("category")
df["Senior Management"]=df["Senior Management"].astype("bool")
df.sort_values("First Name",inplace=True)
df.head()
```

```python
(df["First Name"].duplicated(keep="first")).head(10)
```

```python
(df["First Name"].duplicated(keep="last")).head()#last instance
```

```python
df.head(10)
```

```python
(df[df["First Name"].duplicated(keep=False)]).head(10)#keeps all duplicates values
```

```python
        mask=~df["First Name"].duplicated(keep=False)#selects only unique rows TILDA
```

```python
df[mask]
```

```python
df["First Name"].duplicated(keep=False).head()#makes all the repeated vals as duplicate
```

# The .drop_duplicate()-sorted-on DF inplace

```python
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time",])
df["Gender"]=df["Gender"].astype("category")
df["Senior Management"]=df["Senior Management"].astype("bool")
df.sort_values("First Name",inplace=True)
df.head()
```

```python
len(df)
```

```python
len(df.drop_duplicates())
```

```python
(df.drop_duplicates(subset=["First Name"],keep="last"))
```

```python
(df.drop_duplicates(subset=["First Name"],keep=False)).head()
```

```python
df.drop_duplicates(subset=["Team"],keep=False)#no unique team value
```

```python
(df.drop_duplicates(subset=["Team","First Name"])).head()
#delets only first name and team matches in both rows
```
