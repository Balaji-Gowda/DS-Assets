# The .agg() method

```python
df=pd.read_csv("fortune1000.csv",index_col=["Rank"])
g=df.groupby(["Sector"])#groupby
df.head()
```

```python
g.mean()
```

```python
g.agg({"Revenue":"sum",
      "Profits":"sum",
      "Employees":"mean"})######to perform various operations on various columns
```

```python
g.sum()
```

```python
g.agg(["size","sum","mean"])###multile opns on all colms
```

```python
g.agg({"Revenue":["sum","mean"],
      "Profits":"sum",
      "Employees":"mean"})######to perform multiple operations on various columns
```

# Iterating thriug a groups in group obj

```python
df=pd.read_csv("fortune1000.csv",index_col=["Rank"])
g=df.groupby(["Sector"])#groupby
df.head()
```

```python
g["Profits"].max()#no entire row is showing
```

```python
opn=pd.DataFrame(columns=df.columns)#creating empty DF to store max or min values and their corresodng rows

opn
```

```python
g=df.groupby(["Sector"])#groupby
for sector,data in g:
    output=data.nlargest(1,"Revenue") 
    opn=opn.append(output)
    
```

```python
opn
```

```python
opn2=pd.DataFrame(columns=df.columns)#creating empty DF to store max or min values and their corresodng rows

opn2
```

```python
g2=df.groupby(["Location"])#groupby loc
for city,data2 in g2:
    output2=data2.nlargest(2,"Revenue") 
    opn2=opn2.append(output2)##########process is finding first row in each location grouped with max revenue

    
```

```python
opn2.sort_values(by="Location")
```
