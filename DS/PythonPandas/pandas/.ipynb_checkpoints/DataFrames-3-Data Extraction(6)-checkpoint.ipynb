# .nsmallest() .nlargest()

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
bond.sort_values("Box Office",ascending=False).head()
```

```python
bond.nlargest(3,columns="Box Office")     #1
```

```python
bond.sort_values("Box Office",ascending=False).tail()
```

```python
bond.nsmallest(3,columns="Box Office",keep="first")
```

```python
bond["Box Office"].nsmallest(3)   #2
```

```python
bond.loc["Licence to Kill","Box Office"]=275.2
```

```python
bond["Box Office"].nsmallest(3,keep="last")#2 #last reps smallest value last occured comes first in DF
```

```python
bond["Box Office"].nsmallest(3,keep="first")#2#last reps smallest value last occured comes first in DF
```

```python
bond["Box Office"].nsmallest(1,keep="all")#2#all reps all smallest values of equal level
```

```python
bond.sort_values("Box Office",ascending=False).tail()
```

# DF with Where()

```python
 bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
mask=bond["Actor"]=="Sean Connery"
bond[mask]
```

```python
    bond.where(mask)#Unmatching rows are null
```

```python
mask2=bond["Box Office"]>900
bond.where(mask2)
```

```python
bond.where(mask & mask2)#both condns are not satisfied
```

# DF with query method-  no spaces in column names

# A review of .apply() mtd on single colms

```python
 bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
def conv_tostr_add_milins(num):
    return str(num)+" Millions"
bond["Box Office"]
```

```python
bond["Box Office"].apply(conv_tostr_add_milins)
```

```python
bond["Budget"].apply(conv_tostr_add_milins)
```

```python
colms=["Box Office","Budget","Bond Actor Salary"]
for col in colms:
    bond[col]=bond[col].apply(conv_tostr_add_milins)
```

```python
bond.head()
```

# The .apply() method with single row values

```python
 bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
def good_movie(row):
    actor=row[1]
    budget=row[4]
    if actor=="Pierce Brosnan":
        return "The Best"
    elif actor=="Roger Moore" and budget>40:
        return "Enjoyable"
    else:
        return "I have no clue"
```

```python
bond.apply(good_movie,axis="columns")
```
