# Creat bargraphs to show counts

```python
bb=data.DataReader("BB","yahoo","2007-01-01","2020-12-31")
bb.head(3)
```

```python
bb["Close"]
```

```python
def perfmce(val):
    if val>35:
        return "Steller"
    elif val>18 and val <=35:
        return "Satisfactory"
    else:
        return "Poor"
```

```python
bb["Close"].apply(perfmce).value_counts()
```

```python
plt.style.use("dark_background")
bb["Close"].apply(perfmce).value_counts().plot()
```

```python
bb.plot()
```

```python
bb["Close"].apply(perfmce)
```

```python
plt.style.use("dark_background")
bb["Close"].apply(perfmce).value_counts().plot(kind="bar")
```

```python
plt.style.use("ggplot")
bb["Close"].apply(perfmce).value_counts().plot(kind="bar")
```

```python
plt.style.use("ggplot")
bb["Close"].apply(perfmce).value_counts().plot(kind="barh").plot(kind="barh")
```

# Creating pie Charts

```python
bb=data.DataReader("BB","yahoo","2007-01-01","2020-12-31")
bb.head(3)
```

```python
bb["Close"].mean()
```

```python
def mean(val):
    if val>=29.452764073094578:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("dark_background")
bb["Close"].apply(mean).value_counts().plot(kind="pie")
```

```python
plt.style.use("dark_background")
bb["Close"].apply(mean).value_counts().plot(kind="pie",legend=True)
```

```python
bb.plot(ylabel="new Y axis")
```
