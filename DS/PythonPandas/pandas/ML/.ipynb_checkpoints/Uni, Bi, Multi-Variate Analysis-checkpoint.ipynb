```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("iris.csv").drop(columns='Unnamed: 0')
```

```python
df.head()
```

```python
df.shape
```

# Univariate Analysis

```python
df.head()
```

```python
df.loc[df['species']=='setosa']
```

```python
df_setosa=df[df['species']=='setosa']#three diffreent species in different dfs
```

```python
df_virginica=df[df['species']=='virginica']
```

```python
df_versicolor=df[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepalLength'],np.zeros_like(df_setosa['petalLength']),'*')
plt.plot(df_virginica['sepalLength'],np.zeros_like(df_virginica['petalLength']),'*')
plt.plot(df_versicolor['sepalLength'],np.zeros_like(df_versicolor['petalLength']),'*')
plt.xlabel("sepalLength")
plt.show()
```

```python
plt.plot(df_setosa['sepalLength'],np.zeros_like(df_setosa['petalLength'])+0,'*')
plt.plot(df_virginica['sepalLength'],np.zeros_like(df_virginica['petalLength'])+1,'*')
plt.plot(df_versicolor['sepalLength'],np.zeros_like(df_versicolor['petalLength'])+2,'*')
plt.xlabel("sepalLength")
plt.show()
```

# Bivariate Analysis

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,"sepalLength","sepalWidth").add_legend()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,"petalLength","sepalWidth").add_legend()
```
