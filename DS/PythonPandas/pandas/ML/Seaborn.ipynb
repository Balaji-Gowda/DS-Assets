# Seaborn Tutorials

### Distribution plots - for continuous values like float, int
- distplot
- joinplot
- pairplot

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tip=pd.read_csv("tips.csv")
tip.head()
```

# Correlation with heatmap

```python
tip.dtypes
```

```python
tip.corr()
```

```python
sns.heatmap(tip.corr())
```

## Univariate/BIvariate

```python
#

sns.jointplot(x='tip',y='total_bill',data=tip,kind='hex')
```

```python
#regression-pdf
sns.jointplot(x='tip',y='total_bill',data=tip,kind='reg')
```

### pair plot
- A pair plot is also known as scatterplot in which one variable in same data row is matched with another variable's value like this: pairs plots are just elaborataion on this , showing all variables paired with all other variables


```python
sns.pairplot(tip)
```

```python
#classification
sns.pairplot(tip,hue='sex')
```

```python
#classification
sns.pairplot(tip,hue='smoker')
```

```python
tip["smoker"].value_counts()
```

### Dist Plot
- Its is used to check the distribution of the columns feature
- histogram creations

```python
#kde is true == density in Y axis
sns.distplot(tip["tip"])
```

```python
#kernal density estimation kde==count
sns.distplot(tip["tip"],kde=False)
```

# Categorical PLots

### Plots used to visualise categorical values:
- boxplot
- violinplot
- countplot
- barplot

```python
tip=pd.read_csv("tips.csv")
tip.head()
```

```python
## CountPlot
#Show the counts of observations in each categorical bin using bars.

sns.countplot('sex',data=tip)
```

```python
## CountPlot
#Show the counts of observations in each categorical bin using bars.

sns.countplot('smoker',data=tip)
```

```python
## CountPlot
#Show the counts of observations in each categorical bin using bars.

sns.countplot('day',data=tip)
```

```python
## CountPlot
#Show the counts of observations in each categorical bin using bars.

sns.countplot(y='time',data=tip)
```

# BarPlots

```python
##two axes needed

tip=pd.read_csv("tips.csv")
tip.head()
```

```python
sns.barplot(y='sex',x='total_bill',data=tip)
```

```python
sns.barplot(y='day',x='total_bill',data=tip)
```

```python
sns.barplot(y='smoker',x='total_bill',data=tip)
```

```python
sns.barplot(y='tip',x='time',data=tip)
```

# Box Plots

```python
#Draw a box plot to show distributions with respect to categories
sns.boxplot(x='sex',y='total_bill',data=tip)
```

```python
sns.boxplot(x='day',y='total_bill',data=tip,palette='rainbow')
```

```python
#this haas arger count so above box plot is in that range



m1= tip['total_bill']>=15
m2=tip['total_bill']>5
tip[m1 & m2].value_counts().sum()      # Number of rows
```

```python
tip.info()
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tip)
```

### To plot all columns which are int or float in boxplots

```python
sns.boxplot(data=tip,orient='h')#orient
```

```python
sns.boxplot(data=tip,orient='v')
```

# Violin plot

```python
sns.violinplot(data=tip,orient='h')#Draw a combination of boxplot and kernel density estimate.

#==joint plot
```

```python
#Draw a combination of boxplot and kernel density estimate.

sns.violinplot(x='total_bill',y='day',data=tip,orient='h',palette='rainbow')
```
